```python
#code is used to make all columns heading small letter and space as _ so it will not make any confuion in sql
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create columns age group
labels=['Young Adults','Adults','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

```python
#Frequency mapping in machine learning is a technique used to transform input data into a higher-dimensional space, making it easier to analyze or classify
#purchase frequency days colum creation
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df
```
